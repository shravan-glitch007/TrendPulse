```python
import pandas as pd
```

```python
df = pd.read_json("trends_20260406.json")

df.head()
```

```python
df.info()
df.shape
```

```python
# Remove duplicates
df = df.drop_duplicates(subset="post_id")

# Remove empty titles
df = df[df["title"].notnull()]

# Reset index
df = df.reset_index(drop=True)

df.shape
```

```python
df["collected_at"] = pd.to_datetime(df["collected_at"])
```

```python
df["category"].value_counts()
```

```python
df.to_csv("cleaned_trend.csv", index=False)
```
